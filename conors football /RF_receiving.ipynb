```python
data = df2.values
X = data[:, 0:0]
y = data[:, 0]
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
label_encoder.fit(y)
encoded_y = label_encoder.transform(y)
```

```python
one_hot_list = []
for label, original_class in zip(encoded_y, y):
    print('Player: ' + str(original_class))
    one_hot_list.append(original_class)
df2["one_hot"] = one_hot_list
```

```python
df_labels = df2.drop(columns=["Player","AV","Receiving Yards","Receptions","Receiving TDs"])
df_labels.head()
```

```python
df_labels["Round"].head()
```

```python
import numpy as np

labels = np.array(df_labels["Round"])

feature_list = list(features.columns)

features = np.array(features)

```

```python
from sklearn.model_selection import train_test_split

train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size = 0.25, random_state = 42)
```

```python
print('Training Features Shape:', train_features.shape)
print('Training Labels Shape:', train_labels.shape)
print('Testing Features Shape:', test_features.shape)
print('Testing Labels Shape:', test_labels.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification

rf = RandomForestClassifier(n_estimators = 1000, random_state = 42)
rf.fit(train_features, train_labels);
```

```python
predictions = rf.predict(test_features)

errors = abs(predictions - test_labels)


print('Mean Absolute Error:', round(np.mean(errors), 2))
```

```python
mape = 100 * (errors / test_labels)

accuracy = 100 - np.mean(mape)
print('Accuracy:', round(accuracy, 2), '%.')
```

```python
from sklearn.tree import export_graphviz
import pydot

tree = rf.estimators_[5]

from sklearn.tree import export_graphviz
import pydot

tree = rf.estimators_[5]

export_graphviz(tree, out_file = 'tree.dot', feature_names = feature_list, rounded = True, precision = 1)

(graph, ) = pydot.graph_from_dot_file('tree.dot')

graph.write_png('rectree.png')
```
