```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np 
```

```python
offense_load = "../Offense/clean_offense.csv"
offense_df = pd.read_csv(offense_load)

defense_load = "../Defense/clean_defense.csv"
defense_df = pd.read_csv(defense_load)

NFL_load = "nfl_draft.csv"
NFL_df1 = pd.read_csv(NFL_load)
NFL_df1 = NFL_df1.replace(np.nan, 0)
```

```python
NFL_year = NFL_df1["Year"]
NFL_round = NFL_df1["Rnd"]
NFL_pick = NFL_df1["Pick"]
NFL_team = NFL_df1["Tm"]
NFL_player = NFL_df1["Player"]
NFL_position = NFL_df1["Pos"]
NFL_age = NFL_df1["Age"]
NFL_1stteam_allpro = NFL_df1["AP1"]
NFL_probowl = NFL_df1["PB"]
NFL_careerAV = NFL_df1["CarAV"]
NFL_games = NFL_df1["G"]
NFL_completions = NFL_df1["Cmp"]
NFL_attempts = NFL_df1["Pass_Att"]
NFL_passyards = NFL_df1["Pass_Yds"]
NFL_passtd = NFL_df1["Pass_TD"]
NFL_interception = NFL_df1["Pass_Int"]
NFL_rushattempts = NFL_df1["Rush_Att"]
NFL_rushyards = NFL_df1["Rush_Yds"]
NFL_rushtd = NFL_df1["Rush_TDs"]
NFL_Rec = NFL_df1["Rec"]
NFL_Rec_Yards = NFL_df1["Rec_Yds"]
NFL_rectd = NFL_df1["Rec_Tds"]
NFL_tackles = NFL_df1["Tkl"]
NFL_interceptions = NFL_df1["Def_Int"]
NFL_sacks = NFL_df1["Sk"]
NFL_college = NFL_df1["College/Univ"]



NFL_data_df = pd.DataFrame({
    
    "Year": NFL_year,
    "Round": NFL_round,
    "Pick": NFL_pick,
    "Team": NFL_team,
    "Player": NFL_player,
    "Pos": NFL_position,
    "Age": NFL_age,
    "All Pro": NFL_1stteam_allpro,
    "Pro Bowl": NFL_probowl,
    "AV": NFL_careerAV,
    "Games": NFL_games,
    "Pass Completions": NFL_completions,
    "Pass Attempts": NFL_attempts,
    "Pass Yards": NFL_passyards,
    "Pass TDs": NFL_passtd,
    "Interceptions": NFL_interception,
    "Rush Attempts": NFL_rushattempts,
    "Rush Yards": NFL_rushyards,
    "Rush TDs": NFL_rushtd,
    "Receptions": NFL_Rec,
    "Receiving Yards": NFL_Rec_Yards,
    "Receiving TDs": NFL_rectd,
    "Tackles": NFL_tackles,
    "Picks": NFL_interceptions,
    "Sacks": NFL_sacks 
    
})


NFL_df = NFL_data_df[["Year", "Round", "Pick", "Player", "Pos", "Age", "Team", "All Pro", "Pro Bowl", "AV", "Games", "Pass Completions", "Pass Attempts", "Pass Yards", "Pass TDs", "Interceptions", "Rush Attempts", "Rush Yards", "Rush TDs", "Receptions", "Receiving Yards", "Receiving TDs", "Tackles", "Picks", "Sacks"]]
NFL_df
```

```python
O_df = offense_df.drop(columns=['College', 'School', 'AV', 'Rk', 'Pos'])
offensive_df = O_df.replace(np.nan, 0)
offensive_df.head()
```

```python
offensiveNFL_df = pd.merge(offensive_df, NFL_df, how='inner', on=['Player', 'Year', 'Round'])
offensiveNFL_df.replace(np.nan, 0)
```

```python
offensiveNFL_df.to_csv("offensiveNFL_Data.csv", header=True, index=True)
```

```python
D_df = defense_df.drop(columns=['College', 'School', 'AV', 'Rk', 'Pos'])
defensive_df = D_df.replace(np.nan, 0)
defensive_df
```

```python
defensiveNFL_df = pd.merge(defensive_df, NFL_df, how='inner', on=['Player', 'Year', 'Round'])
defensiveNFL_df.replace(np.nan, 0)
```
